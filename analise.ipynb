```python
#importanto bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importando a base de dados e visualizando
base = pd.read_csv(r'cancelamentos.csv')
display(base)
```

```python
#verificando informaÃ§Ãµes da base (checando valores nulos)

base.info()
```

```python
#deletando valores nulos e verificando as informaÃ§Ãµes

base =  base.dropna()
base.info()
```

```python
#deletando a coluna CustomerID pois nÃ£o sera relevante para a analise

base = base.drop(columns='CustomerID')
display(base)
```

```python
#entendendo total de cancelamentos
print(base['cancelou'].value_counts())
print(base['cancelou'].value_counts(normalize=True).map("{:.1%}".format))
```

```python
#gerando graficos para entender os cancelamentos baseado em cada uma das colunas 

for coluna in base.columns:
    if coluna != "cancelou": 
        plt.figure(figsize=(10, 6))
        for label in base["cancelou"].unique():
            subset = base[base["cancelou"] == label]
            plt.hist(subset[coluna], bins=30, alpha=0.5, label=str(label))

        plt.xlabel(coluna)
        plt.title(f'{coluna} X cancelamento')
        plt.legend(title='Cancelou')
        plt.show()
```

```python
#analisando cada grafico, 4 chamam atenÃ§Ã£o: forma de pagamento, ligaÃ§Ãµes no call center e dias de atraso.
#Notei que todos os clientes que assinavam o plano mensal cancelaram - Clientes que ligam a partir de 5 vezes para call center cancelam  - Clientes que atrasam o pagamento apartir de 21 dias cancelam 

#Removendo da base todos os clientes com o plano mensal 
base = base[base["duracao_contrato"]!="Monthly"]

#Removendo da base todos os clientes com mais de 4 ligaÃ§Ãµes no call center
base = base[base["ligacoes_callcenter"]< 5]

# Removendo da base todos os clientes a partir de 21 dias de atraso no pagamento
base = base[base["dias_atraso"]<=20]

# Removendo da base todos os clientes com de 50 anos
base = base[base["idade"]< 50]

display(base["cancelou"].value_counts())
display(base["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```
